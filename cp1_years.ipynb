```python
df_1920 = df[:9]
df_1930 = df[9:19]
df_1940 = df[19:29]
df_1950 = df[29:39]
df_1960 = df[39:49]
df_1970 = df[49:59]
df_1980 = df[59:69]
df_1990 = df[69:79]
df_2000 = df[79:89]
df_2010 = df[89:99]

df_2010
```

```python
df_1920 = df_1920.mean()
df_1930 = df_1930.mean()
df_1940 = df_1940.mean()
df_1950 = df_1950.mean()
df_1960 = df_1960.mean()
df_1970 = df_1970.mean()
df_1980 = df_1980.mean()
df_1990 = df_1990.mean()
df_2000 = df_2000.mean()
df_2010 = df_2010.mean()

df_2010
```

```python
df_2 = pd.concat([df_1920, df_1930, df_1940, df_1950, df_1960, df_1970, df_1980, df_1990, df_2000, df_2010], axis = 1)


df_2
```

```python
df_2.columns = ['1920s', '1930s','1940s','1950s','1960s','1970s','1980s','1990s','2000s','2010s']
df_2
```

```python
df_3 = df_2.loc[["danceability"],:]
df_4 = df_2.loc[["energy"],:]
df_5 = df_2.loc[["popularity"],:]
df_6 = df_2.loc[["tempo"],:]
df_7 = df_2.iloc[4,:]
```

```python
import seaborn as sb
plt.rcParams["figure.figsize"] = (30,30)
sb.heatmap(df.corr(),
           annot = True, #ì‹¤ì œ ê°’ í™”ë©´ì— ë‚˜íƒ€ë‚´ê¸°
           cmap = 'Greens', #ìƒ‰ìƒ
           vmin = -1, vmax=1 , #ì»¬ëŸ¬ì°¨íŠ¸ ì˜ì—­ -1 ~ +1
          )
```

```python
fig = plt.figure(figsize=(10,10))
fig.set_facecolor('white')
 
plt.scatter(
    x = df['year'], 
    y = df['danceability'], 
    s= df['popularity'], 
    c='green', 
    alpha=0.6, 
    edgecolors="white", 
    linewidth=2);
 
plt.yscale('log')
plt.xlabel("Year")
plt.ylabel("Danceability")
plt.title("Music Trends by Year")
plt.grid(True, which='major', linestyle='--', linewidth=0.5)
plt.show() #ê°€ë¡œ year, #ì„¸ë¡œ danceability # particleí¬ê¸° popularity
```

```python
fig = plt.figure(figsize=(5,5))
fig.set_facecolor('white')
 
plt.scatter(
    x = df['tempo'], 
    y = df['danceability'], 
    s= df['energy'], 
    c='black', 
    alpha=1, 
    edgecolors="white", 
    linewidth=2);

plt.yscale("log")
plt.xlabel("tempo")
plt.ylabel("danceability")
plt.title("Danceability")
plt.grid(True, which='major', linestyle='--', linewidth=0.5)
plt.show()#ê°€ë¡œ tempo, #ì„¸ë¡œ danceability # particleí¬ê¸° energy
```
